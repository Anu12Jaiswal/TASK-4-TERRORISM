```python
#Distribution of Fatalities by Attack Type 
plt.figure(figsize=(12, 6))
data.boxplot(column='nkill', by='attacktype1_txt')
plt.xlabel('Attack Type')
plt.ylabel('Number of Kills')
plt.title('Distribution of Fatalities by Attack Type')
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution of Attack Types by Region
cross_tab = pd.crosstab(data['region_txt'], data['attacktype1_txt'])
cross_tab.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.title('Distribution of Attack Types by Region')
plt.legend(title='Attack Type')
plt.show()
```

```python
#Number of Attacks by Country
plt.figure(figsize=(12, 6))
data['country_txt'].value_counts().head(10).plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.title('Top 10 Countries with Highest Number of Attacks')
plt.show()
```
